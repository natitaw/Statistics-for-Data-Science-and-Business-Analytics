# Categorical Data

- Frequency Distribution
- Bar Charts
- Pie Charts
- Pareto Diagrams

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
cities = ['NY', 'LA', 'SF']
freq = [12.327, 17.129, 19.923]
```

### Frequency Distribution

```python
fig, ax = plt.subplots()

y_pos = np.arange(len(cities))
width = 0.0

# Create bars
rects1 = ax.bar(y_pos, freq)

ax.set_ylabel('Cities')
ax.set_title('Frequency Distribution', color='orange')

ax.set_xticks(y_pos + width / 2)
ax.set_xticklabels(cities, color='orange')


# Create names on the x-axis
#plt.xticks(y_pos, cities)

plt.yticks(color='orange')

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%.2f' % (height),
                ha='center', va='bottom')
        
autolabel(rects1)
```

## Pie Chart

```python
norm = [(x/sum(freq))*100 for x in freq]

f, ax = plt.subplots()



#ax[0,i].set_title(f'{titles[i]}', color='black')
ax.pie(norm, labels=cities, autopct='%1.2f%%', 
            shadow=True, startangle=90)

plt.yticks(color='orange')
```

## Pareto Distr

```python
from matplotlib.ticker import PercentFormatter

rel_freq = []
val = 0
for item in norm:
    rel_freq.append(item+val)
    val+=item

fig, ax = plt.subplots()
ax.bar(cities, freq, color="C0")
ax2 = ax.twinx()
ax2.plot(cities, rel_freq, color="C1", marker="D", ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="orange")
ax2.tick_params(axis="y", colors="orange")
ax.set_xticklabels(cities, color='orange')

plt.show()
```
